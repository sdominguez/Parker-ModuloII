#KNN

Ejemplo de uso del algoritmo KNN para un problema de clasificación. Iris-dataset

```
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

#Cargar datos

```
iris = load_iris()
x= pd.DataFrame(iris.data,columns=iris.feature_names)
y = pd.Categorical.from_codes(iris.target, iris.target_names)

x.info()
```

```
#Preparacion de datos de entrenaimento y prueba
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

#Entrenamiento

```
#Creación del modelo
#KNN_Classifier = KNeighborsClassifier(n_neighbors = 6, p = 2, metric='minkowski')
KNN_Classifier = KNeighborsClassifier(n_neighbors = 4, metric='euclidean')

#Aplicar aprendizaje
KNN_Classifier.fit(x_train, y_train)

```

Predicción

```
pred_test = KNN_Classifier.predict(x_test)
print(pred_test)
```

```
# obtener las probabilidades de la predicción (opcional)
probabilidades_prediccion = KNN_Classifier.predict_proba(x_test)
print(iris.target_names)
print(probabilidades_prediccion)
```

#Evaluación

```
#verficar la pecisión del modelo
acc = accuracy_score(y_test, pred_test)

acc
```

```
print(classification_report(y_test, pred_test))
```

```
cm = confusion_matrix(y_test, pred_test)
cm
```
